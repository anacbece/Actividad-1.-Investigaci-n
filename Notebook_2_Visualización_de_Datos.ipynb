```python
#Importación de librerías y carga del dataset

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

sns.set(style="whitegrid")

# Cargar dataset Titanic
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
#Gráficos Básicos con Matplotlib

#Histograma de edades
plt.figure(figsize=(8,5))
plt.hist(titanic['age'].dropna(), bins=30)
plt.title("Distribución de Edad de los Pasajeros")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.show()

#Gráfico de barras: Sobrevivientes por sexo
plt.figure(figsize=(7,5))
titanic.groupby('sex')['survived'].sum().plot(kind='bar')
plt.title("Sobrevivientes por Sexo")
plt.xlabel("Sexo")
plt.ylabel("Cantidad")
plt.show()
```

```python
#Gráficos Avanzados con Seaborn

#Boxplot de edad por clase
plt.figure(figsize=(8,5))
sns.boxplot(data=titanic, x="class", y="age")
plt.title("Distribución de Edad por Clase")
plt.show()


#Heatmap de correlación
plt.figure(figsize=(10,6))
sns.heatmap(titanic.corr(numeric_only=True), annot=True, cmap="Blues")
plt.title("Matriz de Correlación")
plt.show()

#Countplot: pasajeros por clase
plt.figure(figsize=(7,5))
sns.countplot(data=titanic, x="class")
plt.title("Número de Pasajeros por Clase")
plt.show()

#Scatterplot (edad vs tarifa)
plt.figure(figsize=(8,5))
sns.scatterplot(data=titanic, x="age", y="fare", hue="survived")
plt.title("Relación Edad vs Tarifa")
plt.show()
```

```python
#Bonus: Visualizaciones Interactivas con Plotly

#Gráfico interactivo: edad vs tarifa
fig = px.scatter(
    titanic,
    x="age",
    y="fare",
    color="survived",
    title="Edad vs Tarifa (Interactivo)"
)
fig.show()

#Barras interactivas: pasajeros por clase
fig = px.bar(
    titanic.groupby("class")["survived"].count().reset_index(),
    x="class",
    y="survived",
    title="Pasajeros por Clase (Interactivo)"
)
fig.show()

```
